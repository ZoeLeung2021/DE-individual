### XF Driver stats 2015-2022

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import re
```

```python
# Define function for getting nascar xf series racing driver stats data from api (from 2015-2022)
def driver_data_xf(year):
    base_url='https://api.sportradar.us/nascar-ot3/xf/drivers/'
    year=year
    url_2='/drivers.json?api_key=qu7qgb45bmkbfcwkrnxjh7ga'
    url=base_url+year+url_2
    r = requests.get(url)
    json_data = r.json()
    df = pd.DataFrame(json_data['drivers'])
    return df
```

```python
# Get 7 dataframe of nascar xf series racing driver stats data
driver_data_xf_2022=driver_data_xf('2022')
driver_data_xf_2021=driver_data_xf('2021')
driver_data_xf_2020=driver_data_xf('2020')
```

```python
driver_data_xf_2019=driver_data_xf('2019')
```

```python
driver_data_xf_2018=driver_data_xf('2018')
driver_data_xf_2017=driver_data_xf('2017')
driver_data_xf_2016=driver_data_xf('2016')
driver_data_xf_2015=driver_data_xf('2015')
```

```python
# Defined a function to drop irrelevant and ambiguous defined columns
def drop_columns(df):
    df=df.drop(['status','residence','track_type_splits','track_splits','race_splits','children','hobbies','college'],axis=1)
    return df
```

```python
# Drop the irrelevant and ambiguous columns for 7 tables
driver_data_xf_2022=drop_columns(driver_data_xf_2022)
driver_data_xf_2020=drop_columns(driver_data_xf_2020)
driver_data_xf_2019=drop_columns(driver_data_xf_2019)
driver_data_xf_2018=drop_columns(driver_data_xf_2018)
driver_data_xf_2017=drop_columns(driver_data_xf_2017)
driver_data_xf_2016=drop_columns(driver_data_xf_2016)
driver_data_xf_2015=drop_columns(driver_data_xf_2015)
```

```python
# Year 2021 data did not contain college column, therefore define another function for it
def drop_columns_without_college(df):
    df=df.drop(['status','residence','track_type_splits','track_splits','race_splits','children','hobbies'],axis=1)
    return df
```

```python
# Year 2021 data did not contain college column, therefore do it seperately
driver_data_xf_2021=drop_columns_without_college(driver_data_xf_2021)
```

```python
# Define a function to add a column called series in order to make clarification at the later merging stage 
# with driver standings data

def add_series(df,series):
    df['series']=series
    return df
```

```python
# Apply the add series function to the 7 tables
driver_data_xf_2022=add_series(driver_data_xf_2022,'xf')
driver_data_xf_2021=add_series(driver_data_xf_2021,'xf')
driver_data_xf_2020=add_series(driver_data_xf_2020,'xf')
driver_data_xf_2019=add_series(driver_data_xf_2019,'xf')
driver_data_xf_2018=add_series(driver_data_xf_2018,'xf')
driver_data_xf_2017=add_series(driver_data_xf_2017,'xf')
driver_data_xf_2016=add_series(driver_data_xf_2016,'xf')
driver_data_xf_2015=add_series(driver_data_xf_2015,'xf')
```

```python
# Concatenate the seven years into one big table for xf series driver stats data and drop duplicated rows 
driver_xf=pd.concat([driver_data_xf_2022,driver_data_xf_2021, driver_data_xf_2020,driver_data_xf_2019,driver_data_xf_2018,driver_data_xf_2017,driver_data_xf_2016,driver_data_xf_2015], axis=0)
driver_xf=driver_xf.drop_duplicates(subset=['id'],keep='first')
```

### Rookie year information -- from wikipedia recent driver list

Since Debut year is an important factor for the analysis at the later stage and this column contains a large amount of missing values, this information will be obtained mainly through wikipedia pages. As there is a wikipedia page called 'list of nascar driver' which collected recent nascar drivers from all racing series with their debut year information. It would be the first resource to use for filling the missing value.

```python
# Specify url: url
url = 'https://en.wikipedia.org/wiki/List_of_NASCAR_drivers'

# Package the request, send the request and catch the response: r
r = requests.get(url)

# Extract the response as html: html_doc
html_doc = r.text

# Create a BeautifulSoup object from the HTML: soup
soup = BeautifulSoup(html_doc)

soup
```

```python
# Use regular expression to get the driver name in the wikepedia page
soup_body = str(soup)                                                                             
driver=re.findall(r'(?:.*">(.*)</a></span></span></span>|<td>.*href=".*>(.*)</a>\n</td>\n<td>)',soup_body)
```

```python
# Define a function to combine the two format result together
def combine(list_name):
    driver_list=[]
    for i in range(len(list_name)):
        driver_list.append(''.join(list(list_name[i])))
    return driver_list
```

```python
# Combine the two different format result
driver=combine(driver)
```

```python
driver
```

```python
# Defined a function to get the index of a driver in the list 
def get_index_positions(list_of_elems, element):
    index_pos_list = []
    index_pos = 0
    while True:
        try:
            index_pos = list_of_elems.index(element, index_pos)
            index_pos_list.append(index_pos)
            index_pos += 1
        except ValueError as e:
            break
    return index_pos_list
```

### The rest missing value for rookie year -- each driver's wikipedia page

```python
# Define the web scraping function for each driver
def web_scrap(name):
    base_url = 'https://en.wikipedia.org/wiki/'
    query = name
    url=base_url+query
    r = requests.get(url)
    html_doc = r.text
    soup = BeautifulSoup(html_doc)
    soup_body = str(soup)
    first_race=re.findall(r'First\srace.*?title="(\d{4}?).*NASCAR\s(.*?)">\d{4}</a>',soup_body)
    return first_race
```

```python
# List of drivers name that still contains missing debut year information
rookie_na=driver_stats_new[driver_stats_new.rookie_year.isna()]
rookie_na_list=rookie_na.full_name.to_list()
rookie_na_list
```

```python
# Web scarping and store the debut year information in the list
rookies=[]
for i in rookie_na_list:
    rookies.append(web_scrap(i))
```

```python
# Make a df
rookies_fill = pd.DataFrame(
    {'full_name': rookie_na_list,
     'first_race': rookies})
```

```python
# Format cleaning for later stage processing
rookies_fill['first_race'] =  rookies_fill['first_race'].apply(lambda x: str(x).replace('[','').replace(']','')) 
rookies_fill['first_race'] = rookies_fill.first_race.replace('',np.nan,regex = True)
rookies_fill=rookies_fill.set_index('full_name').apply(lambda x: x.str.split('[)]').explode()).reset_index()
rookies_fill.drop(rookies_fill[rookies_fill.first_race==''].index,inplace=True)
```
