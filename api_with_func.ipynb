### XF Driver stats 2015-2022

```python
import requests
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
# Define function for getting nascar xf series racing driver stats data from api (from 2015-2022)
def driver_data_xf(year):
    base_url='https://api.sportradar.us/nascar-ot3/xf/drivers/'
    year=year
    url_2='/drivers.json?api_key=qu7qgb45bmkbfcwkrnxjh7ga'
    url=base_url+year+url_2
    r = requests.get(url)
    json_data = r.json()
    df = pd.DataFrame(json_data['drivers'])
    return df
```

```python
# Get 7 dataframe of nascar xf series racing driver stats data
driver_data_xf_2022=driver_data_xf('2022')
driver_data_xf_2021=driver_data_xf('2021')
driver_data_xf_2020=driver_data_xf('2020')
```

```python
driver_data_xf_2019=driver_data_xf('2019')
```

```python
driver_data_xf_2018=driver_data_xf('2018')
driver_data_xf_2017=driver_data_xf('2017')
driver_data_xf_2016=driver_data_xf('2016')
driver_data_xf_2015=driver_data_xf('2015')
```

```python
# Defined a function to drop irrelevant and ambiguous defined columns
def drop_columns(df):
    df=df.drop(['status','residence','track_type_splits','track_splits','race_splits','children','hobbies','college'],axis=1)
    return df
```

```python
# Drop the irrelevant and ambiguous columns for 7 tables
driver_data_xf_2022=drop_columns(driver_data_xf_2022)
driver_data_xf_2020=drop_columns(driver_data_xf_2020)
driver_data_xf_2019=drop_columns(driver_data_xf_2019)
driver_data_xf_2018=drop_columns(driver_data_xf_2018)
driver_data_xf_2017=drop_columns(driver_data_xf_2017)
driver_data_xf_2016=drop_columns(driver_data_xf_2016)
driver_data_xf_2015=drop_columns(driver_data_xf_2015)
```

```python
# Year 2021 data did not contain college column, therefore define another function for it
def drop_columns_without_college(df):
    df=df.drop(['status','residence','track_type_splits','track_splits','race_splits','children','hobbies'],axis=1)
    return df
```

```python
# Year 2021 data did not contain college column, therefore do it seperately
driver_data_xf_2021=drop_columns(driver_data_xf_2021)
```

```python
# Define a function to add a column called series in order to make clarification at the later merging stage 
# with driver standings data

def add_series(df,series):
    df['series']=series
    return df
```

```python
# Apply the add series function to the 7 tables
driver_data_xf_2022=add_series(driver_data_xf_2022,'xf')
driver_data_xf_2021=add_series(driver_data_xf_2021,'xf')
driver_data_xf_2020=add_series(driver_data_xf_2020,'xf')
driver_data_xf_2019=add_series(driver_data_xf_2019,'xf')
driver_data_xf_2018=add_series(driver_data_xf_2018,'xf')
driver_data_xf_2017=add_series(driver_data_xf_2017,'xf')
driver_data_xf_2016=add_series(driver_data_xf_2016,'xf')
driver_data_xf_2015=add_series(driver_data_xf_2015,'xf')
```

```python
# Concatenate the seven years into one big table for xf series driver stats data and drop duplicated rows 
driver_xf=pd.concat([driver_data_xf_2022,driver_data_xf_2021, driver_data_xf_2020,driver_data_xf_2019,driver_data_xf_2018,driver_data_xf_2017,driver_data_xf_2016,driver_data_xf_2015], axis=0)
driver_xf=driver_xf.drop_duplicates(subset=['id'],keep='first')
```
